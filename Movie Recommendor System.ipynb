```python
import pandas as pd
import numpy as np
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Read the movie dataset

movies=pd.read_csv(r"C:\Users\Admin\Downloads\archive (13)\\movies.csv")
movies.head()
```

```python
# Drop rows with missing values and reset the index

movies = movies.dropna()
movies = movies.reset_index(drop=True)
```

```python
# Create a TfidfVectorizer object to convert the movie genres into a matrix of TF-IDF features

tfidf = TfidfVectorizer(stop_words='english')
tfidf
```

```python
# Replace the '|' separator between genres with a space and fit_transform the genres into a matrix of TF-IDF features

movies['genres'] = movies['genres'].str.replace('|', ' ')
tfidf_matrix = tfidf.fit_transform(movies['genres'])
```

```python
# Compute the cosine similarity between all pairs of movies based on their genre features

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
# Create a Pandas Series object with the movie titles as index and their corresponding row numbers as values

indices = pd.Series(movies.index, index=movies['title']).drop_duplicates()
```

```python
# Define a function that takes a movie title and number of recommendations as input and returns the top n similar movies based on their genre features

def get_recommendations(title, n):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:n+1]
    movie_indices = [i[0] for i in sim_scores]
    return movies['title'].iloc[movie_indices]
```

```python
# Get user input for movie title and number of recommendations and call the get_recommendations function to get the top n similar movies

t = input("Enter movie title: ")
n = int(input("Enter number of recommendations: "))
```

```python
# Call the get_recommendations function to get the top n similar movies based on the user input

recs = get_recommendations(t, n).reset_index(drop=True)
```

```python
# Reset the index of the DataFrame and add 1 to each index value

recs.index += 1
```

```python
# Convert the Series object to a DataFrame and reset the index

recs = recs.to_frame().reset_index()
```

```python
# Rename the columns of the DataFrame to 'SRNo.' and 'Title'
recs.columns = ['SRNo.', 'Title']
```

```python
# To prompt user for input

user_movie = int(input("UserID: "))
num_recommend = int(input("Number of recommendations (N): "))
similar_users_thresh = int(input("Threshold for similar users (k): "))
```

```python
# To apply filter ratings by target user

user_ratings = ratings[ratings['userId'] == user_movie]
user_ratings.head()
```

```python
# To calculate  similarity between users

sim_user_rat = ratings.pivot(index='userId', columns='movieId', values='rating').fillna(0)
user_similarity = cosine_similarity(sim_user_rat)
sim_user_rat.head()
```

```python
user_similarity
```

```python
# Getting top K similar users

similar_users_index = user_similarity[user_movie - 1].argsort()[::-1][1:similar_users_thresh + 1]
similar_users_index
```

```python
#  Movies rated by similar users

similar_users_ratings = sim_user_rat.iloc[similar_users_index]
similar_users_movies = similar_users_ratings.columns[sim_user_rat.loc[user_movie] == 0]
similar_users_ratings.head()
```

```python
similar_users_movies
```

```python
# Calculate movie recommendations

recommendations1 = pd.DataFrame(similar_users_ratings[similar_users_movies].mean().sort_values(ascending=False))
recommendations1.columns = ['Avg_Rating']
recommendations1.index.name = 'movieId'
recommendations1.head()
```

```python
# To merge with movies dataset to get movie titles

recommendations1 = pd.merge(recommendations1, movies, on='movieId').head(num_recommend)
recommendations1.head()
```
